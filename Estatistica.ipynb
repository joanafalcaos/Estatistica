```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
planilha = pd.read_excel('base.xlsx')
```

```python
# Tabela de Frequência para Sexo
sexo_freq = planilha['SEXO'].value_counts()
print("Tabela de Frequência para Sexo:\n", sexo_freq)

# Gráfico de Barras para Sexo
plt.figure(figsize=(8, 6))
sns.countplot(x='SEXO', data=planilha, palette='Set2')
plt.title('Distribuição dos Alunos por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Número de Alunos')
plt.xticks(rotation=0)
plt.show()

# Tabela de Frequência para Turno
turno_freq = planilha['TURNO'].value_counts()
print("Tabela de Frequência para Turno:\n", turno_freq)

# Gráfico de Barras para Turno
plt.figure(figsize=(8, 6))
sns.countplot(x='TURNO', data=planilha, palette='Set2')
plt.title('Distribuição dos Alunos por Turno')
plt.xlabel('Turno')
plt.ylabel('Número de Alunos')
plt.xticks(rotation=0)
plt.show()

# Tabela de Frequência para Modalidade do Curso
modalidade_freq = planilha['MODALIDADE'].value_counts()
print("Tabela de Frequência para Modalidade do Curso:\n", modalidade_freq)

# Gráfico de Barras para Modalidade do Curso
plt.figure(figsize=(8, 6))
sns.countplot(x='MODALIDADE', data=planilha, palette='Set2')
plt.title('Distribuição dos Alunos por Modalidade do Curso')
plt.xlabel('Modalidade do Curso')
plt.ylabel('Número de Alunos')
plt.xticks(rotation=0)
plt.show()

# Tabela de Frequência para Categoria Acadêmica
categoria_freq = planilha['CategoriaAdm'].value_counts()
print("Tabela de Frequência para Categoria Acadêmica:\n", categoria_freq)

# Gráfico de Barras para Categoria Acadêmica
plt.figure(figsize=(8, 6))
sns.countplot(x='CategoriaAdm', data=planilha, palette='Set2')
plt.title('Distribuição dos Alunos por Categoria Acadêmica')
plt.xlabel('Categoria Acadêmica')
plt.ylabel('Número de Alunos')
plt.xticks(rotation=0)
plt.show()

# Tabela de Contingência para Sexo e Modalidade do Curso
contingencia_sexo_modalidade = pd.crosstab(planilha['SEXO'], planilha['MODALIDADE'])
print("Tabela de Contingência para Sexo e Modalidade do Curso:\n", contingencia_sexo_modalidade)

# Gráfico de Barras Empilhadas para Sexo e Modalidade do Curso
contingencia_sexo_modalidade.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Pastel1')
plt.title('Distribuição dos Alunos por Sexo e Modalidade do Curso')
plt.xlabel('Sexo')
plt.ylabel('Número de Alunos')
plt.xticks(rotation=0)
plt.show()

# Análise Descritiva das Notas por Modalidade do Curso
notas_desc_presencial = planilha[planilha['MODALIDADE'] == 1][['NOTA_GERAL', 'NOTA-FORM_GER', 'NOTA_COMPESPE']].describe()
notas_desc_ead = planilha[planilha['MODALIDADE'] == 0][['NOTA_GERAL', 'NOTA-FORM_GER', 'NOTA_COMPESPE']].describe()

print("Estatísticas Descritivas das Notas para Cursos Presenciais:\n", notas_desc_presencial)
print("Estatísticas Descritivas das Notas para Cursos EAD:\n", notas_desc_ead)

# BoxPlot das Notas por Modalidade do Curso
plt.figure(figsize=(12, 8))
sns.boxplot(data=planilha[['NOTA_GERAL', 'NOTA-FORM_GER', 'NOTA_COMPESPE']], palette='Pastel1')
plt.title('Distribuição das Notas')
plt.xlabel('Tipo de Nota')
plt.ylabel('Nota')
plt.show()

# BoxPlot das Notas por Modalidade do Curso
plt.figure(figsize=(12, 8))
sns.boxplot(x='MODALIDADE', y='NOTA_GERAL', data=planilha, palette='Pastel1')
plt.title('Distribuição das Notas Gerais por Modalidade do Curso')
plt.xlabel('Modalidade do Curso')
plt.ylabel('Nota Geral')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='MODALIDADE', y='NOTA-FORM_GER', data=planilha, palette='Pastel1')
plt.title('Distribuição das Notas de Formação Geral por Modalidade do Curso')
plt.xlabel('Modalidade do Curso')
plt.ylabel('Nota de Formação Geral')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='MODALIDADE', y='NOTA_COMPESPE', data=planilha, palette='Pastel1')
plt.title('Distribuição das Notas do Componente Específico por Modalidade do Curso')
plt.xlabel('Modalidade do Curso')
plt.ylabel('Nota do Componente Específico')
plt.show()
```
